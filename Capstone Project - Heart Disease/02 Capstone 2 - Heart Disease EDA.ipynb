## Visualization 

```python
#Age of people with heart disease 
heart.age.plot(kind ='hist', bins=25)
plt.xlabel("Age")
plt.ylabel('Frequency')   
plt.title('Value Frequency for Age ')
```

```python
#Frequency of chest pain correlated with four types
heart.cp.plot(kind ='hist', bins=25)
plt.xlabel("Chest Pain Type 4 types")
plt.ylabel('Frequency')   
plt.title('Value Frequency for Chest Pain ')
```

```python
#Resting Blood Pressure for the 302 peoples
heart.trestbps.plot(kind ='hist', bins=25)
plt.xlabel("Resting Blood Pressure")
plt.ylabel('Frequency')   
plt.title('Value Frequency for Resting Blood Pressure ')
```

```python
#Cholesterol frequency for the each patient 
heart.chol.plot(kind ='hist', bins=25)
plt.xlabel("Cholesterol")
plt.ylabel('Frequency')   
plt.title('Value frequency for Cholesterol ')
```

```python
# Maximum Heart Rate Achieved (thalach) 
heart.thalach.plot(kind ='hist', bins=25)
plt.xlabel("Max Heart Rate Acheived")
plt.ylabel('Frequency')
plt.title('Value frequency for Max Heart Rate Achieved')
```

```python
# oldpeak (ST depression induced by exercise relative to rest)
heart.oldpeak.plot(kind ='hist', bins=25)
plt.xlabel("Oldpeak")
plt.ylabel('Frequency')   
plt.title('Frequency: depression induced by exercise relative to rest')
```

```python
# thal: 3 = normal; 6 = fixed defect; 7 = reversable defect 
heart.thal.plot(kind ='hist', bins=25)
plt.xlabel("Thal")
plt.ylabel('Frequency') 
plt.title('Value frequency for Thal')
```

```python
#Target 
heart.target.plot(kind ='hist', bins=25)
plt.xlabel("Target")
plt.ylabel('Frequency')   
plt.title('Value frequency for Target')
```

Above histograms have illistrated the basic frequncy of each of our numerical features. With each of the enumerical values consistant within the perspective feature, we see that when looking at the thal feature, the perspective values are dependend on normal, fixed defect, and reversable defet, 160 are less than normal.
